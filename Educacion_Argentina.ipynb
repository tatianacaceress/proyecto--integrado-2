```python
# prompt: df.to_csv("datos_alumnos.csv", index=False)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuración de datos
materias = [
    "Lengua y Literatura", "Matemáticas", "Ciencias Naturales",
    "Física", "Química", "Historia", "Geografía", "Inglés",
    "Educación Física", "Filosofía", "Sociología", "Artes", "Biología"
]

# Generar datos aleatorios
np.random.seed(42)  # Semilla para reproducibilidad
n_registros = 100

data = {
    "Apellido": np.random.choice(["Gomez", "Perez", "Lopez", "Martinez", "Rodriguez"], n_registros),
    "Nombre": np.random.choice(["Juan", "Maria", "Luis", "Ana", "Sofia", "Pedro"], n_registros),
    "Nro_DNI": np.random.randint(30000000, 45000000, n_registros),
    "Materia Escolar": np.random.choice(materias, n_registros),
    "Nota 1er Trimestre": np.random.randint(1, 11, n_registros),
    "Nota 2do Trimestre": np.random.randint(1, 11, n_registros),
    "Nota 3er Trimestre": np.random.randint(1, 11, n_registros),
    "Faltas a clases": np.random.randint(0, 16, n_registros),
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Exportar a CSV (no se modifica este bloque)
df.to_csv("datos_alumnos.csv", index=False)
print("Archivo CSV generado: datos_alumnos.csv")

# El resto del código permanece igual
# Análisis Estadístico
materia_stats = df.groupby("Materia Escolar").agg({
    "Nota 1er Trimestre": "mean",
    "Nota 2do Trimestre": "mean",
    "Nota 3er Trimestre": "mean",
    "Faltas a clases": "mean"
}).round(2)

materia_stats["Promedio General"] = materia_stats[[
    "Nota 1er Trimestre", "Nota 2do Trimestre", "Nota 3er Trimestre"
]].mean(axis=1)

materia_stats = materia_stats.sort_values("Promedio General", ascending=False)

# Gráfico
fig, ax1 = plt.subplots(figsize=(12, 6))

# Promedio General
ax1.bar(materia_stats.index, materia_stats["Promedio General"], color="skyblue", label="Promedio General")
ax1.set_ylabel("Promedio General")
ax1.set_xlabel("Materias")
ax1.tick_params(axis="x", rotation=45)

# Faltas a Clases
ax2 = ax1.twinx()
ax2.plot(materia_stats.index, materia_stats["Faltas a clases"], color="orange", label="Faltas Promedio", marker="o")
ax2.set_ylabel("Faltas Promedio")

# Leyendas
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.title("Promedio General y Faltas Promedio por Materia")
plt.tight_layout()
plt.show()
```
